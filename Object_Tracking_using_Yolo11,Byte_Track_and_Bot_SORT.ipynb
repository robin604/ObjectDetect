# Detection

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11m.pt")

# Perform object detection on video
results = model("/content/CityUtdR.mp4", save=True)
```

```python
#results = model("/content/p.mp4",classes=0, save=True)  - #Classes=0 to detect person class
```

# Track

```python
#Bot Sort
results = model.track(source="/content/CityUtdR.mp4", persist=True, save=True)
```

```python
# Byte Track
results = model.track("/content/CityUtdR.mp4", persist=True, show=True, classes=0, tracker="bytetrack.yaml")
```
