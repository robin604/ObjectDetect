```python
import os
import csv

# Define the path to the folder containing label text files
labels_folder = '/content/yolov7/runs/detect/exp2/labels'
# Define the output CSV file path
output_csv = '/content/yolov7/detections.csv'

# CSV header: you can customize this if you wish to convert normalized values to pixel values later.
header = ['image', 'class_id', 'x_center', 'y_center', 'width', 'height', 'confidence']

# Open the CSV file for writing
with open(output_csv, mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(header)

    # Iterate over all text files in the labels folder
    for txt_file in os.listdir(labels_folder):
        if txt_file.endswith('.txt'):
            # Derive the image file name (assuming it has the same base name with a .jpg extension)
            image_name = txt_file.replace('.txt', '.jpg')
            file_path = os.path.join(labels_folder, txt_file)

            with open(file_path, 'r') as f:
                for line in f:
                    # Split the line to extract detection details
                    values = line.strip().split()
                    # Write a row to the CSV: image name followed by detection details
                    writer.writerow([image_name] + values)

print("Detections exported to:", output_csv)
```

Step 1: Export Your YOLOv7 Model to ONNX Format
Before using Bolt application, you need to export your trained YOLOv7 model from Google Colab to ONNX format:

```python
# Install required packages
!pip install onnx
!pip install onnxruntime

# Now load and export your model
import torch
from models.experimental import attempt_load

# Load your trained model
model = attempt_load('runs/train/exp/weights/best.pt', map_location=torch.device('cpu'))

# Export to ONNX
torch.onnx.export(
    model,
    torch.zeros((1, 3, 640, 640)),
    'yolov7.onnx',
    opset_version=12,
    input_names=['images'],
    output_names=['output']
)

# Download the ONNX model
from google.colab import files
files.download('yolov7.onnx')
```

To create a CSV file with your image paths for use with the YOLOv7 inference app. Since your images are in Google Drive, you'll need to create a CSV file that contains the URLs to these images.

```python
import os
import pandas as pd
from google.colab import drive

# Mount Google Drive if not already mounted
drive.mount('/content/drive')

# Path to your images folder
image_folder = '/content/drive/MyDrive/ObjectDetect/InferanceImages/'

# Get list of image files
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Create a function to get shareable link for each file
def get_shareable_link(file_path):
    # This requires manual steps to share files and get links
    # For demonstration, we'll create a placeholder
    return f"https://drive.google.com/uc?id=YOUR_FILE_ID_FOR_{os.path.basename(file_path)}"

# Create DataFrame with image URLs
data = []
for img_file in image_files:
    # In a real scenario, you would replace this with actual shareable links
    data.append({'image_url': get_shareable_link(os.path.join(image_folder, img_file))})

# Create CSV file
df = pd.DataFrame(data)
df.to_csv('inference_images.csv', index=False)

# Download the CSV file
from google.colab import files
files.download('inference_images.csv')
```
